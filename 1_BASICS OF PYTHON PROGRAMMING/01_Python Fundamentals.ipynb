# 4. Keywords & Identifiers

> What is python interpreter ?
- There is a concept in programming which is called compilation. Compilation is basically a process where you compile your code to **binary** language.In python the language is english , python is a high level programming language. But the problem is that your machine where you write code and execute it is basicaly a **chip**, and chip can not understand english(python syntax). Chip only can understand binary numbers [0, 1]. And so for execution the code on your machine at first you had to convert your english code into binary [0, 1] pattern . And this convertion process is called compilation. And this compilation done by **compiler** or **interpreter**

- Low level programming language like: `c`, `jaba` use **compiler**
- High level programming language like: `python`, `JabaScript` use **interpreter** (line by line execution ) 

## Keywords
- And your **Compiler** or **interpreter** need help to undertand how to convert english to binary. There are certain word which are reserved by the **compiler** and **interpreter** who use those word for iteself to understand program.They are called **keyword**.There are 32 keyword in python

# 6. Type Conversion
**Type convertion is a way to convert your datatype to one datatype to another datatype.Type conversion doesn't create permanent changes**
- There are two types of type convertion in python
    - **Implicit** : where type convertion internally done by interpreter
    - **Explicit** : where user define the type of data
        - str(), int(), float(), oct(), dict()

```python
# Implicit type conversion
print(5+5.6)

print(type(5), type(5.6))
```

```python
# Implicit always not work because of some retriction
print(type(5),type(5.6))

print(4 + '4')
```

```python
# Explicit
# str -> int

# int to str
str(5)

# float
float(4)
print(4 + int('4'))
```
